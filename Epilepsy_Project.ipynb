Identifying missing values

```python
import pyspark.sql.functions as F
df = data.agg(*[F.count(F.when(F.isnull(c), c)).alias(c) for c in data.columns])
```

```python
display(df)
```

Outliers in the data
Data points above and below 1.5*(interquartile distance) are being removed

```python
bounds = {c: dict(zip(["q1","q3"], data.approxQuantile(c,[0.25,0.75],0))) for c in data.columns}
```
