Identifying missing values

```python
import pyspark.sql.functions as F
df = data.agg(*[F.count(F.when(F.isnull(c), c)).alias(c) for c in data.columns])
```

```python
display(df)
```

```python
cols = list(set(data.columns)-set(["class"]))
```

Outliers in the data
Data points above and below 1.5*(interquartile distance) are being removed

```python
bounds = {c: dict(zip(["q1","q3"], data.approxQuantile(c,[0.25,0.75],0))) for c in cols}
```

```python
print(bounds)
```

```python
for c in bounds:
    iqr = bounds[c]['q3'] - bounds[c]['q1']
    bounds[c]['lower'] = bounds[c]['q1'] - (1.5*iqr)
    bounds[c]['upper'] = bounds[c]['q3'] + (1.5*iqr)
print(bounds)
```

```python
import pyspark.sql.functions as F
combined2 = data.select("*",*[F.when(F.col(c).between(bounds[c]['lower'],bounds[c]['upper']), F.col(c)).otherwise(F.lit(None)).alias(c+"1") for c in cols])
```

```python
display(combined2)
```

```python
display(combined2.agg(*[F.count(F.when(F.isnull(c), c)).alias(c) for c in combined2.columns]))
```

```python
combined_wo = combined2.dropna()
```

```python
display(combined_wo.agg(*[F.count(F.when(F.isnull(c), c)).alias(c) for c in combined_wo.columns]))
```

Selecting original columns

```python
combined_wo = combined_wo.select([c for c in combined_wo.columns if c in data.columns])
```

Building model

```python
from pyspark.ml.feature import VectorAssembler
inputcolumns = list(combined_wo.columns)
vectorAssembler = VectorAssembler(inputCols = inputcolumns, outputCol = 'features')
```

```python
vdata1 = vectorAssembler.transform(combined2)
vdata1 = vdata1.select(['features', 'class'])
display(vdata1)
```

Splitting data into train and test

```python
splits = vdata1.randomSplit([0.7,0.3], seed = 24)
train_df = splits[0]
test_df = splits[1]
```

```python
from pyspark.ml.feature import MinMaxScaler
scaler = MinMaxScaler(inputCol = "features", outputCol = "scaledFeatures")
scalerModel = scaler.fit(train_df.select("features"))
scaledtrain = scalerModel.transform(train_df)
scaledtest = scalerModel.transform(test_df)
```

```python
from pyspark.ml.classification import RandomForestClassifier
```

```python
rf = RandomForestClassifier(featuresCol = "features", labelCol = 'class', numTrees = 6, featureSubsetStrategy='auto', maxDepth=5)
```

```python
rfmodel = rf.fit(scaledtrain)
predictions = rfmodel.transform(scaledtest)
```

```python
print(predictions)
```

```python
predictions.select('class', 'prediction', 'probability').show(5, False)
```

```python
# Create a confusion matrix
predictions.groupBy('class', 'prediction').count().show()
```

```python
# Calculate the elements of the confusion matrix
TN = predictions.filter('prediction = 0 AND class = prediction').count()
TP = predictions.filter('prediction = 1 AND class = prediction').count()
FN = predictions.filter('prediction = 0 AND class != prediction').count()
FP = predictions.filter('prediction = 1 AND class != prediction').count()
```

```python
# Accuracy measures the proportion of correct predictions
accuracy = (TN + TP) / (TN + TP + FN + FP)
print(accuracy)
```
